```python
dept = { "department_id": [1,2,3],
         "department":    ["Engineering", "Sales", "Marketing"]
}


staff =  {
            "name":      ["Alice", "Bob", "Charlie", "Diana", "Erin"],
            "age":       [25,       30,    None,     35,       28],
            "active":    [True,     False, False,     True,     False],
            "category":  ["A",      "B",   None,     "A",      "C"],
            "salary":    [50000,    75000, None,     60000,    80000],
            "years_exp": [2,        5,     0,        8,        3],
            "dept_id":   [1,        2,     2,        1,        3]
        }



#Polars Expression
pl_T_STAFF = pl.DataFrame(staff)
pl_T_STAFF.filter( pl.col("name").is_null() )

#Narhwals filterexpression
nwpl_T_STAFF = nw.from_native(pl_T_STAFF)
nwpl_T_STAFF.filter( nw.col("name").is_null() )

#Table: Staff
NAME =      nw.col("name")
AGE =       nw.col("age")
ACTIVE =    nw.col("active")
CATEGORY =  nw.col("category")
SALARY =    nw.col("salary")
YEARS_EXP = nw.col("years_exp")
DEPT_ID =   nw.col("dept_id")

# Table: Department
DEPARTMENT_ID = nw.col("department_id")
DEPARTMENT =    nw.col("department")

#Derived Columns
SALARY_PER_YEAR_EXP = SALARY / YEARS_EXP

# Tables
pl_T_STAFF = pl.DataFrame(staff)
pl_T_DEPT = pl.DataFrame(dept)

ib_T_STAFF = ibis_con.create_table( "staff",obj=pl_T_STAFF, overwrite=True).mutate("SALARY_PER_YEAR_EXP" = SALARY_PER_YEAR_EXP)
ib_T_DEPT = ibis_con.create_table(  "dept", obj=pl_T_DEPT, overwrite=True)

nwpl_T_STAFF = nw.from_native(pl_T_STAFF)
nwpl_T_DEPT = nw.from_native(pl_T_DEPT)

nwib_T_STAFF = nw.from_native(ib_T_STAFF)
nwib_T_DEPT = nw.from_native(ib_T_DEPT)

```
